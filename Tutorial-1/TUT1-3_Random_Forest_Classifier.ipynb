```python
X_train = np.loadtxt(open("/kaggle/input/aist4010-spring2024-a0/X_train.csv","rb"), delimiter=",", skiprows=0)
y_train = np.loadtxt(open("/kaggle/input/aist4010-spring2024-a0/y_train.csv","rb"), delimiter=",", skiprows=0)
```

```python
print('X_train:\n', X_train[0:3])
print('y_train:\n', y_train[0:3])
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
clf.fit(X_train, y_train)
```

```python
X_test = np.loadtxt(open("/kaggle/input/aist4010-spring2024-a0/X_test.csv","rb"), delimiter=",", skiprows=0)
y_pred = clf.predict(X_test)
```

```python
import pandas as pd
output = {'Predicted':y_pred}
output_df = pd.DataFrame(output)
```
